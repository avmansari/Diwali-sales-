```python
# Import necessary libraries for data manipulation and visualization
import numpy as np # Linear algebra library
import pandas as pd # Data manipulation and analysis library
import matplotlib.pyplot as plt # Plotting library
%matplotlib inline
import seaborn as sns # Statistical data visualization library (built on matplotlib)
```

```python
# Load the dataset from a CSV file into a pandas DataFrame
# encoding='unicode_escape' is used to handle potential encoding issues in the file
df = pd.read_csv(r'/content/drive/MyDrive/Diwali Sales Data 1.csv', encoding='unicode_escape')
```

```python
# --- Data Cleaning ---
# This section focuses on cleaning the data by handling missing values and unnecessary columns.
```

```python
# Display information about the DataFrame, including data types and non-null counts
df.info()
```

```python
# Drop the 'Status' and 'unnamed1' columns as they contain only null values
df.drop(['Status', 'unnamed1'], axis=1 , inplace=True)
```

```python
# Display information about the DataFrame again to confirm the columns have been dropped
df.info()
```

```python
# Drop rows with any null values
df.dropna(inplace=True)
```

```python
# Display information about the DataFrame again to confirm null values have been removed
df.info()
```

```python
# Verify that there are no null values left in the DataFrame
pd.isnull(df).sum()
```

```python
# Change the data type of the 'Amount' column from float to integer
df['Amount'] = df['Amount'].astype(int)
```

```python
# --- Exploratory Data Analysis (EDA) ---
# This section explores the data through visualizations and summary statistics to gain insights.
```

```python
# Create a count plot of the 'Gender' column to visualize the distribution of buyers by gender
ax = sns.countplot(x = 'Gender', data = df)

# Add labels to the bars to show the exact counts
for bars in ax.containers:
    ax.bar_label(bars)

# Display the plot
plt.title('Distribution of Buyers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Buyers')
plt.show()
```

```python
# Group data by 'Gender' and calculate the total 'Amount' spent by each gender
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Display the total sales by gender
print("Total sales by gender:")
display(sales_gen)
```

```python
# Create a bar plot to visualize the total sales amount by gender
sns.barplot(x ='Gender',y ='Amount', data = sales_gen, palette=['blue', 'pink'])

# Add titles and labels
plt.title('Total Sales Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')

# Display the plot
plt.show()
```

```python
# Create a count plot to visualize the distribution of buyers by 'Age Group' and 'Gender'
ax=sns.countplot(data = df,x = 'Age Group', hue = 'Gender')

# Add labels to the bars to show the exact counts
for bars in ax.containers:
  ax.bar_label(bars)

# Add titles and labels
plt.title('Distribution of Buyers by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Number of Buyers')

# Display the plot
plt.show()
```

```python
# Group data by 'State', calculate the total 'Orders' for each state, and get the top 10
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending= False).head(10)

# Set the figure size for better readability of the bar plot
sns.set(rc={'figure.figsize':(15,5)})

# Create a bar plot to visualize the top 10 states by number of orders
sns.barplot(data =sales_state, x = 'State', y = 'Orders')

# Add titles and labels
plt.title('Top 10 States by Number of Orders')
plt.xlabel('State')
plt.ylabel('Number of Orders')

# Display the plot
plt.show()
```

```python
# Group data by 'State', calculate the total 'Amount' spent in each state, and get the top 10
foramt = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending= False).head(10)

# Set the figure size for better readability of the bar plot
sns.set(rc={'figure.figsize':(18,5)})

# Create a bar plot to visualize the top 10 states by sales amount
sns.barplot(data = foramt, x = 'State', y = 'Amount')

# Add titles and labels
plt.title('Top 10 States by Sales Amount')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')

# Display the plot
plt.show()
```

```python
# Create a count plot to visualize the distribution of buyers by 'Marital_Status'
ax = sns.countplot(data = df, x = 'Marital_Status')

# Set the figure size for better readability of the bar plot
sns.set(rc={"figure.figsize":(10,8)}) # Increased figure size for better label visibility

# Add labels to the bars to show the exact counts
for bars in ax.containers:
  ax.bar_label(bars)

# Add titles and labels
plt.title('Distribution of Buyers by Marital Status')
plt.xlabel('Marital Status (0: Unmarried, 1: Married)')
plt.ylabel('Number of Buyers')
plt.xticks([0, 1], ['Unmarried', 'Married']) # Label the x-axis ticks

# Display the plot
plt.show()
```

```python
# Create a count plot to visualize the distribution of buyers by 'Marital_Status' and 'Gender'
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Gender')

# Set the figure size for better readability of the bar plot
sns.set(rc={"figure.figsize":(6,4)}) # Adjusted figure size

# Add labels to the bars to show the exact counts
for bars in ax.containers:
  ax.bar_label(bars)

# Add titles and labels
plt.title('Distribution of Buyers by Marital Status and Gender')
plt.xlabel('Marital Status (0: Unmarried, 1: Married)')
plt.ylabel('Number of Buyers')
plt.xticks([0, 1], ['Unmarried', 'Married']) # Label the x-axis ticks

# Display the plot
plt.show()
```

```python
# Create a count plot to visualize the distribution of buyers by 'Occupation'
ax= sns.countplot(data = df, x= 'Occupation' )

# Set the figure size for better readability of the bar plot
sns.set(rc={'figure.figsize':(20,5)})

# Add labels to the bars to show the exact counts
for bars in ax.containers:
  ax.bar_label(bars)

# Add titles and labels
plt.title('Distribution of Buyers by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Buyers')

# Rotate x-axis labels for better readability if there are many occupations
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Display the plot
plt.show()
```

```python
# Group data by 'Occupation', calculate the total 'Amount' spent by each occupation, and get the top 10
octbyamt = df.groupby(['Occupation'])['Amount'].sum().sort_values(ascending=False).head(10).reset_index()

# Set the figure size for better readability of the bar plot
sns.set(rc={'figure.figsize':(10,7)})

# Create a bar plot to visualize the top 10 occupations by sales amount
sns.barplot(x = 'Occupation',y = 'Amount', data=octbyamt)

# Add titles and labels
plt.title('Top 10 Occupations by Sales Amount')
plt.xlabel('Occupation')
plt.ylabel('Total Sales Amount')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Display the plot
plt.show()
```
